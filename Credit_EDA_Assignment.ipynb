# Missing values

```python
# checking the percentage of null values in columns

null_data_percentage = application_data.isnull().sum() * 100/len(application_data)
null_data_percentage.sort_values(ascending=True)
```

```python
# checking the percentage of null values greater than 35%

major_missing_data_columns = null_data_percentage[null_data_percentage >= 35]
major_missing_data_columns
```

```python
#num of major null values in columns

len(major_missing_data_columns)
```

```python
# dropping the major missing values in columns 
application_data_df = application_data.drop(columns=major_missing_data_columns.index)
```

```python
# shape of updated data set
application_data_df.shape
```

```python
"""Now, we done with null values in columns which are greater than 35% 
Lets, come to the null values which are less than 20%"""

minor_missing_values=null_data_percentage[(null_data_percentage <= 20) & (null_data_percentage  > 0)].sort_values(ascending=False)
minor_missing_values
```

# Checking and Converting dtype for Final Dataset

```python
application_data_df.nunique().sort_values(ascending=True)
```

```python
# we have noticed that till "ORGANIZATION_TYPE " columns are categorical type
# lets get their index and convert them into categorical dtype

application_data_df.nunique().sort_values().index
```

```python
# Changing dtype to Category
for i in application_data_df.nunique().sort_values().index[:-12]:
    application_data_df[i]=application_data_df[i].astype('category')
```

```python
# SK_ID_CURR is int data type, but it holds id number of customers, and this variable cannot be manipulated
# so we will convert it to object datatype

application_data_df['SK_ID_CURR'] = application_data_df['SK_ID_CURR'].astype('object')
```

```python
# Some columns (DAYS_BIRTH,DAYS_EMPLOYED,DAYS_REGISTRATION,DAYS_ID_PUBLISH) contains negative value
application_data_df[['DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']].describe()
```

```python
# Let's change the negative to positive values using abs() and change the days to years.

days_col = ['DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']
application_data_df[days_col] = application_data_df[days_col].abs()
application_data_df[days_col] = application_data_df[days_col]/365
application_data_df[days_col].describe()
```

```python
# Let's change columns name as well
application_data_df.rename(columns={'DAYS_BIRTH':'Years_Birth','DAYS_EMPLOYED':'Years_Employed','DAYS_REGISTRATION':'Years_Registration','DAYS_ID_PUBLISH':'Years_Id_Publish'},inplace=True)
```

```python
#as we can see that to visualize 4 plots we wrote same code multiple times. so to avoid redundancy, and to save our time,
#we will put the above code in a function and generalize it for our following plots, 
#so that its easy to visualize and saves time.

def diff_plots(variable_name):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(20,6), sharey='all')
    fig.tight_layout(pad=10.0)

    ax1 = sns.countplot(ax=ax1,data=approved_df,x=variable_name)
    ax1.set_title("Refused", fontsize=10)
    ax1.set_ylabel("Number of Loans")
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90)

    ax2 = sns.countplot(ax=ax2,data=refused_df,x=variable_name)
    ax2.set_title("Approved", fontsize=10)
    ax2.set_ylabel("Number of Loans")
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=90)

    ax3 = sns.countplot(ax=ax3,data=canceled_df,x=variable_name)
    ax3.set_title("Canceled", fontsize=10)
    ax3.set_xlabel(variable_name)
    ax3.set_ylabel("Number of Loans")
    ax3.set_xticklabels(ax3.get_xticklabels(),rotation=90)

    ax4 = sns.countplot(ax=ax4,data=unused_df,x=variable_name)
    ax4.set_title("Unused", fontsize=10)
    ax4.set_xlabel(variable_name)
    ax4.set_ylabel("Number of Loans")
    ax4.set_xticklabels(ax4.get_xticklabels(),rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
diff_plots('NAME_FAMILY_STATUS')
```
